```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Classification_Method = ['XGBoost','LDA', 'QDA','KNN (k=9)',
'Decision Tree',
'Random Forest',
'AdaBoost',
'SVC', 'Logistic Regression','Catboost', 'LightGBM']
accuracy2 = [0.9463257465,0.8864452702,
0.7368880122, 0.8770189061,
0.9043976819,
0.9364257029,
0.9275428752,
0.8722988803, 0.8872859217,
0.9450896333,
0.9448325913]

accuracy1 = [0.8896594661,0.8213748234,
0.733479361,
0.8460586655,
0.850184437,
0.8901043854,
0.8805209705,
0.8118963745,
0.8413063588,
0.8925980024,
0.8902671429]
x = np.arange(len(Classification_Method))
width = 0.2
fig, ax = plt.subplots(figsize=(20,6))
ax.set_xticks(x+0.25) #set the current tick locations
plt.xticks(fontsize=15)
ax.set_xticklabels(Classification_Method)#set the current tick labels of the x-axis
rects1 = ax.bar(x+width,accuracy1,width,label='Standard Window')
rects2 = ax.bar(x+(width*2),accuracy2,width,label='Stepping Window')
ax.legend()
ax.set_ylabel('Average Accuracy', fontsize = 18)
ax.set_xlabel('Classifier', fontsize = 18)
ax.set_title('Average Accuracy By Classifier',fontsize = 20)

fig.tight_layout()
```
