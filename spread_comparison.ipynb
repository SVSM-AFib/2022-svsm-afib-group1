```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import patches as mpatches
from sklearn import preprocessing
import wfdb
import copy as cp
import scipy.signal as signal
from scipy import stats
from sklearn import preprocessing
from tqdm import tqdm
import os
import re
import pandas as pd
import pickle
import csv
```

```python
comp_df = pd.read_csv('spread_comparison.csv', index_col=0)
```

```python
labels = comp_df.index.tolist()
x = np.arange(len(labels))
width = 0.35

all_vals = comp_df['All'].str.rstrip('%').astype('float')
prod_vals = comp_df['Productive'].str.rstrip('%').astype('float')

prod_vals
```

```python
fig, ax = plt.subplots(figsize=(18,8), constrained_layout=True)
rects1 = ax.bar(x - width/2, all_vals, width, label='All')
rects2 = ax.bar(x + width/2, prod_vals, width, label='Productive')

ax.set_ylabel('Standard Deviation (%)', fontsize=20)
ax.set_title('Standard Deviations Compared Across Feature Sets', fontsize=24)
ax.set_xticks(x, labels)
ax.tick_params(labelsize=18)
ax.legend(fontsize=16)

ax.bar_label(rects1, padding=3, fontsize=16)
ax.bar_label(rects2, padding=3, fontsize=16)
```
